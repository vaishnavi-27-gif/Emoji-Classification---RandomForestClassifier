Principle component analysis

```python
from sklearn import decomposition
```

Component Assignment for PCA

```python
pca=decomposition.PCA(n_components=24,whiten=True ,random_state=1)
```

Fitting the training set to generate principle components

```python
pca.fit(x_train)
```

Transforming Principle Components

```python
x_train_pca=pca.transform(x_train)
x_test_pca=pca.transform(x_test)
```

```python
x_train_pca.shape
```

```python
x_test_pca.shape
```

Viewing the principle components or eigen faces

Reshape the image back to the original matrix size

```python
eigen=(np.reshape(x[10],(350,350)).astype(np.float64))
```

```python
eigen
```

plotting Images one by one as Subplot

```python
fig=plt.figure(figsize=(30,30))
for i in range(10):
    ax=fig.add_subplot(3,5,i+1,xticks=[],yticks=[])
    ax.imshow(pca.components_[i].reshape(eigen.shape),cmap=plt.cm.bone)
```

Using RandomForestClassifier Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
model.fit(x_train_pca,y_train)
```

Image Prediction

```python
y_pred=model.predict(x_test_pca)
```

```python
y_pred
```

prediction Visualization

```python
for i in (np.random.randint(0,6,6)):
    predicted_images=(np.reshape(x_test[i],(350,350)).astype(np.float64))
    plt.title('predicted label: {0}'.format(y_pred[i]))
    plt.imshow(predicted_images,interpolation='nearest',cmap='gray')
    plt.show()
```
